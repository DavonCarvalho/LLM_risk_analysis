```python
import pandas as pd
```

```python
dataset1=pd.read_csv("news_articles.csv")
dataset2=pd.read_csv("LLM_articles.csv")
```

```python
dataset1.head()
```

```python
dataset2.head()
```

```python
dataset2.rename(columns={"Name":"Title"},inplace=True)
dataset1.rename(columns={"published":"Date Published","link":"URL","source":"Publisher"}, inplace=True)
```

```python
dataset2.head()
```

```python
dataset2.info()
```

```python
dataset2['Date Published']=pd.to_datetime(dataset2['Date Published'])
dataset1['Date Published']=pd.to_datetime(dataset1['Date Published'])
```

```python
dataset2.head()
```

```python
dataset2['Published Month']=dataset2['Date Published'].dt.strftime('%Y-%m')
dataset1['Published Month']=dataset1['Date Published'].dt.strftime('%Y-%m')
```

```python
dataset1.head()
```

```python
dataset1.drop(["description","published_year"],axis='columns')
dataset2.drop("Description",axis="columns")
```

```python
dataset=pd.concat([dataset1,dataset2],axis=0)
dataset.head()
```

```python
dataset.drop("Unnamed: 0",axis="columns",inplace=True)
```

```python
dataset.describe(include="all")
```

```python
dataset.to_csv("news_dataset.csv")
```
